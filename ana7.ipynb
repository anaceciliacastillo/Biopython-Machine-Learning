```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.1, random_state=42)
X_train_scaled = X_train
X_test_scaled = X_test

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Apply the same transformation to test data
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled.shape, X_test_scaled.shape)
print('X_train_scaled contains all the classes [0, 1, 2]:', np.unique(y_train, return_counts=True))
print('X_test_scaled contains all the classes [0, 1, 2]:', np.unique(y_test, return_counts=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import precision_score, recall_score

rf = RandomForestClassifier(max_depth=12, max_samples=0.63)
rf.fit(X_train_scaled, y_train)

train_pred = rf.predict(X_train_scaled)
test_pred = rf.predict(X_test_scaled)

print('Random forest: train accuracy: ', accuracy_score(train_pred, y_train), ' test accuracy: ', accuracy_score(test_pred, y_test))
print("\nClassification Report:")
print(classification_report(y_test, test_pred, target_names=['0', '1', '2']))
print("Confusion Matrix:")
conf_matrix = confusion_matrix(y_test, test_pred)
print(conf_matrix)

precision = precision_score(y_test, test_pred, average=None)
recall = recall_score(y_test, test_pred, average=None)

print("Precision:", precision,"\nRecall:", recall)


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Class 0', 'Class 1', 'Class 2'],
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Feature importance
feature_names = X1.columns
importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Gini Importance': importances}).sort_values('Gini Importance', ascending=False) 
feature_importance_df.to_csv('featureImportance.csv')

print("\nTop 20 Important Features:")
print(feature_importance_df.head(7))

# Plot feature importance
plt.figure(figsize=(25, 8))
sns.barplot(x='Feature', y='Gini Importance', data=feature_importance_df.head(7))
plt.title('Top 7 Feature Importance')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Apply the same transformation to test data
X_test_scaled = scaler.transform(X_test)

logReg = LogisticRegression(max_iter=200, solver='liblinear', penalty='l1', C=0.1, random_state=42)
logReg.fit(X_train_scaled, y_train)

train_pred = logReg.predict(X_train_scaled)
test_pred = logReg.predict(X_test_scaled)

print('Logistic regression:train accuracy: ', accuracy_score(train_pred, y_train), ' test accuracy: ', accuracy_score(test_pred, y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Standardize the data for PCA and t-SNE
# Normalize the data for PCA and t-SNE
X1 = (X1 - X1.min()) / (X1.max() - X1.min()) 

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X1)

# Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X1)

# Plot PCA
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1], label='Thermophiles', alpha=0.5)
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1], label='Psychrophiles', alpha=0.5)
plt.scatter(X_pca[y == 2, 0], X_pca[y == 2, 1], label='Mesophiles', alpha=0.5)
plt.title('PCA Visualization')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()

# Plot t-SNE
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1], label='Thermophiles', alpha=0.5)
plt.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1], label='Psychrophiles', alpha=0.5)
plt.scatter(X_tsne[y == 2, 0], X_tsne[y == 2, 1], label='Mesophiles', alpha=0.5)
plt.title('t-SNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend()

plt.tight_layout()
plt.show()
```
